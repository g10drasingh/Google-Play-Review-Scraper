```python
import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews_all
import plotly.express as px

pd.options.mode.copy_on_write = True
```

```python
from transformers import pipeline

sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")

#Import Sentiment Analysis Model
```

# Get All Reviews

```python
eSewa_reviews = reviews_all(
    'com.f1soft.esewa',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST) # defaults to Sort.MOST_RELEVANT
```

```python
#eSewa_reviews
#check generated reviews
```

```python
df = pd.json_normalize(eSewa_reviews)
```

```python
df['content'] = df['content'].astype('str')

#convert review message to String
```

# Select No of Reviews to Analyse
#Or

```python
dfn=df
```

```python
dfn=dfn[dfn['at']>'2024-03-13 00:00:00']
```

```python
#dfn
# Check Generated Output
```

# Sentiment Analysis

```python
dfn['result'] = dfn['content'].apply(lambda x: sentiment_analysis(x))

#Conduct Sentiment Analysis
```

```python
dfn.head(20)
#Show Result with sentiment score
```

```python
y = sentiment_analysis("I love this!")
```

```python
y[0]['label']
```

```python
y[0]['score']
```

```python
dfn.loc[:,'sentiment'] = dfn['result'].apply(lambda x: (x[0]['label']))
#Creates new coloumn with sentiment value as positive or negative

dfn.loc[:,'score'] = dfn['result'].apply(lambda x: (x[0]['score']))
#Creates new coloumn with sentiment score of each value
```

```python
dfn['sentiment'].value_counts(normalize=True)
#Normalize shows value in %, use() instead to get value in numbers.
```
